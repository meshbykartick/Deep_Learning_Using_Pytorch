## Data Preprocessing

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()# Scaling TheData
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

### Numpy to Pytorch Tensor

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

# A simple Nural Network Without PyTorch Library

```python
class MySimpleNN():

  def __init__(self, X):

    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
learning_rate = 0.1
epochs = 50
```

```python
# create model
model = MySimpleNN(X_train_tensor)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model.forward(X_train_tensor)

  # loss calculate
  loss = model.loss_function(y_pred, y_train_tensor)

  # backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
model.bias
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```

# <font color="lightblue"> Now Defining the same model using PyTorch Packages


```python
import torch
import torch.nn as nn
import torch.optim as optim


X_train_tensor = X_train_tensor.to(torch.float)
y_train_tensor = y_train_tensor.to(torch.float)
X_test_tensor = X_test_tensor.to(torch.float)
y_test_tensor = y_test_tensor.to(torch.float)



# Define the model
class MySimpleNN(nn.Module):
    def __init__(self, input_dim):
        super(MySimpleNN, self).__init__()
        self.linear = nn.Linear(input_dim, 1)

    def forward(self, X):
        z = self.linear(X)
        y_pred = torch.sigmoid(z)
        return y_pred


learning_rate = 0.1
epochs = 50

# Create the model
input_dim = X_train_tensor.shape[1]
model = MySimpleNN(input_dim).to(torch.float)


criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(epochs):

    y_pred = model(X_train_tensor).squeeze()

    loss = criterion(y_pred, y_train_tensor)

    optimizer.zero_grad()
    loss.backward()

    optimizer.step()

    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')

# Model evaluation
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f'Accuracy: {accuracy.item()}')
```
